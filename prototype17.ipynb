```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model

from tensorflow.keras.optimizers import Adam
from keras.models import Model
from keras.models import Input
from keras.layers import Activation
from keras.initializers import RandomNormal
from keras.layers import Concatenate
#from keras_contrib.layers.normalization.instancenormalization import InstanceNormalization
from keras.utils.vis_utils import plot_model
import numpy as np
# example of defining the discriminator model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LeakyReLU
from tensorflow.keras.optimizers import RMSprop
from keras.regularizers import l2
```

```python
# define the standalone generator model
def define_generator(latent_dim=11, n_outputs=11, t="A"):
  init = RandomNormal(mean=0.0, stddev=0.02)
  model = Sequential()
  model.add(Dense(512, activation='relu', kernel_initializer=init, input_dim=latent_dim, name ="input_G_{}".format(t) ))
  model.add(Dropout(0.01))
  model.add(Dense(32,activation='relu',kernel_initializer=init,name = "G_h_layer_{}".format(t)))
  model.add(Dense(n_outputs, activation='linear',name = "output_G_{}".format(t)))
  return model
 
# define the discriminator model
model = define_generator(11)
# summarize the model
# plot the model
plot_model(model, to_file='generator_plot.png', show_shapes=True, show_layer_names=True)
```

```python
 import tensorflow as tf
 def define_discriminator(n_inputs=(11)):
  model = Sequential()
  init = RandomNormal(mean=0.0, stddev=0.02)
  model.add(Dense(n_inputs,kernel_initializer="he_uniform", input_dim=n_inputs, name= "Input_layer_D"))
  model.add(Dropout(0.01))
  model.add(Dense(32,kernel_initializer=init, name  = "Hidden_layer_D"))
  model.add(LeakyReLU(0.2))
  model.add(Dense(1, activation='sigmoid', name = "output_D"))
  # compile model
  
  opt3 = RMSprop(lr=0.00005)
  opt = Adam(lr=0.0002, beta_1=0.5)
  opt1 = tf.keras.optimizers.SGD(lr = 0.0001,decay=1e-6, momentum=0.5, nesterov=True)
  model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
  return model
model1 = define_discriminator(11)
# summarize the model
model1.summary()
# plot the model
plot_model(model1, to_file='generator_plot.png', show_shapes=True, show_layer_names=True)
```

```python
"""
Cette fonction nous oblige en quelque sorte à garder la même taille pour l'entrée et la sortie. 
Si on voulait transformer un vecteur 1x11 en 1x12, lors du calcul la fonction de perte d'identité et le backward cycle 
on aura une erreur, car nous allons alimenté un générateur avec un vecteur (1x11), 
alors qu'il est supposé recevoir un vecteur de taille 1x12 et inversement.
Faudrait trouver une façon de changer la taille de l'input durant l'entrainement, 
mais je n'ai pas trouvé une façon de coder ça sur les forums.



"""


# define a composite model for updating generators by adversarial and cycle loss
def define_composite_model(g_model_1, d_model, g_model_2, i_shape=(11)):
  # le premier générateur entrainable
  g_model_1.trainable = True
# Discriminateur s'entraine tous seul
  d_model.trainable = False
# deuxiéme générateur s'entraine à l'aide d'un deuxiéme modéle composé 
  g_model_2.trainable = False
# Adversial Loss 
  input_gen = Input(shape=i_shape, name = "input_A")  #Input 
  gen1_out = g_model_1(input_gen)   # output = B
  output_d = d_model(gen1_out)     # fake or real D(B)
# Identity function 
  input_id = Input(shape=i_shape ,name = "input_B") #InputB
  output_id = g_model_1(input_id) # same (b)
# forward cycle()
  output_f = g_model_2(gen1_out) # output A
# backward cycle
  gen2_out = g_model_2(input_id) #output A
  output_b = g_model_1(gen2_out) #output B

  model = Model([input_gen, input_id], [output_d, output_id, output_f, output_b])
  #             [X_realA, X_realB],    [y_real, X_realB, X_realA, X_realB]
  
# define optimization algorithm configuration
  opt = Adam(lr=0.0002, beta_1=0.5)
  opt1 = RMSprop(lr=0.00005)
  opt2 = tf.keras.optimizers.SGD(lr=0.01)
#1;1;5;10
 # compile model with weighting of least squares loss and L1 loss
  model.compile(loss=['mse','mae','mae','mae'], loss_weights=[1,2,4,4], optimizer=opt)
  return model
```

```python
i_shape = (11)
# generator: A -> B
g_model_AtoB = define_generator(t="AtoB")
g_model_AtoB._name = "Generateur_A_to_B"
# generator: B -> A
g_model_BtoA = define_generator(t="BtoA")
g_model_BtoA._name = "Generateur_B_to_A"
# discriminator: A -> [real/fake]
d_model_A = define_discriminator()
d_model_A._name = "Discriminateur_A"
# discriminator: B -> [real/fake]
d_model_B = define_discriminator()
d_model_B._name = "Discriminateur_B"


# composite: A -> B -> [real/fake, A]
c_model_AtoBtoA = define_composite_model(g_model_AtoB, d_model_B, g_model_BtoA)
# composite: B -> A -> [real/fake, B]
c_model_BtoAtoB = define_composite_model(g_model_BtoA, d_model_A, g_model_AtoB)

plot_model(c_model_AtoBtoA, to_file='generator_plot.png', show_shapes=True, show_layer_names=True)
#[X_realA, X_realB], [y_real, X_realB, X_realA, X_realB
```

```python

from numpy.random import randint

# select a batch of random samples, returns images and target
def generate_real_samples(dataset, n_samples, patch_shape):
	trainA, trainB = dataset
	ix = randint(0, trainA.shape[0], n_samples)
	X1, X2 = trainA[ix], trainB[ix]
	y = np.ones((n_samples, 1))
	return [X1, X2], y 
```

```python
def generate_fake_samples(g_model, samples, patch_shape):
	X = g_model.predict(samples)
	y = np.zeros((len(X), 1))
	return X, y
```

```python
# train cyclegan models
def train(d_model_A, d_model_B, g_model_AtoB, g_model_BtoA, c_model_AtoB, c_model_BtoA,n,dataset):
  n_batch = 1
  n_steps = 20000

  for i in range(n_steps):
    [X_realA, X_realB], y_real = generate_real_samples(dataset, 32, 1)
    X_fakeA, y_fakeA = generate_fake_samples(g_model_BtoA,X_realB,1)
    X_fakeB, y_fakeB = generate_fake_samples(g_model_AtoB, X_realA,1)
#X_fakeA = update_input_pool(poolA, X_fakeA)
    #X_fakeA = update_input_pool(poolA, X_fakeA)
    #X_fakeB = update_input_pool(poolB, X_fakeB)
#X_fakeB = update_input_pool(poolB, X_fakeB)
    g_loss2, _, _, _, _  = c_model_BtoA.train_on_batch([X_realB, X_realA], [y_real, X_realA, X_realB, X_realA])
    dA_loss1 = d_model_A.train_on_batch(X_realA, y_real)
    dA_loss2 = d_model_A.train_on_batch(X_fakeA, y_fakeA)
# update generator A->B via adversarial and cycle loss
    g_loss1, _, _, _, _ = c_model_AtoB.train_on_batch([X_realA, X_realB], [y_real, X_realB, X_realA, X_realB])
# update discriminator for B -> [real/fake]
    dB_loss1 = d_model_B.train_on_batch(X_realB, y_real)
    dB_loss2 = d_model_B.train_on_batch(X_fakeB, y_fakeB)
# summarize performance    
    print((i+1,"DA_Loss1['loss', 'accuracy'] " ,dA_loss1,"....",dB_loss2,"LOSS_1", g_loss1, "LOSS_2 ",g_loss2))
 
    
```

```python
train(d_model_A, d_model_B, g_model_AtoB, g_model_BtoA, c_model_AtoBtoA, c_model_BtoAtoB, 1,dataset_)
```
